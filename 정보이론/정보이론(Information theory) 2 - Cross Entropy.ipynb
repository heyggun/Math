## Cross Entropy

#### 분포 p의 entropy




- 다음과 같이 기계 P, 기계 Q가 있을 때,
- 기계 P는 A,B,C,D 를 각각 0.25의 확률로 출력
- 기계 Q는 A,B,C,D 를 각각 0.5,0.125,0.125,0.25의 확률로 출력




- 기계 Q에서 entropy 값은 `(0.25x2) x 4 = 2`, 이고 기계 P에서 entropy 값은 `0.5x1 + (0.125x3)x2 + 0.25x2 =1.75` 
- 확률이 랜덤할수록(다 같은 확률로 등장하는 경우가 가장 랜덤함) entropy는 더 높음
- 결국, entropy는 최적화된 전략 하에서 질문개수에 대한 기대값


- 그러나, 만약에 다른 전략을 쓴다면? 전략 Q를 기계 P에 적용한다면? entropy는 아래와 같이 바뀜


- 기존의 1.75던 entropy보다 0.25가 커진 2가 결과값임 
- 이것이 바로 `cross entropy` 임. `Cross Entropy`는 어떤 문제에 대해 `특정 전략을 쓸 때 예상되는 질문 개수에 대한 기댓값`
- 여기서 전략 또한 확률분포로 표현됨. 왜냐하면, 전략 Q는 A,B,C,D의 확률이 모두 0.25 라는 가정에서 나오기 때문임


- 해당 문제(문제 또한 확률분포로 표현됨)에 대한 최적의 전략을 사용할 때 cross entropy 값이 최소가 됨
#### - `확률분포로 된 어떤 문제 p에 대해 확률분포로 된 어떤 전략 q를 사용할 때의 질문개수의 기댓값` : cross entropy

## Cross entropy의 쓰임새



- Pi: 특정 확률에 대한 참값 또는 목표 확률
- qi : 우리가 현재 학습한 확률값
- 여기서는 p = [0.5,0.125,0.125,0.25], q =[0.25,0.25,0.25,0.25]
- 우리가 어떤 qi를 학습하고 있는 상태라면, pi에 가까워질수록 cross entropy값은 작아짐 
- 이산형이 아니라 연속형인 확률 분포에서는 시그마가 아니라 integral이 들어오게 됨



### Binary classification과 cross entropy


- Logistic regression에서 보는 cost function 이지만, cross entropy 식임
- Binary classification에서 0또는 1로 두가지 class를 구분함. 즉, y∈ { 0, 1 }
- 어떤 대상이 1이라고 predict하는 확률 qy=1을 y^, 대상을 0으로 predict하는 확률 qy=0 을 1-y^


- 실제로 어떤 대상이 0 또는 1일 확률(참값)은 py=1, py=0은 각각 y, (1-y)임


#### Cross entropy와 Log loss

- Cross entropy는 log loss로 불리기도 함
- cross entropy를 최소화하는 것은 log likelihood를 최대화 하는 것과 같기 떄문임
- 어떤 데이터가 0 또는 1로 predict 될 확률은 `y^`, `1-y^` 이므로, 이 데이터의 likelihood식은


- y=1 일 때 y^을 최대화시켜야하고, y=0 일 때 1-y^이 최대화 되어야함
- 여기에 log를 씌워보면 지수들이 내려와서 아래와 같은 식이 됨



- 최소화 해야 하는 식은 방금 살펴본 cross entropy와 같은 식임, 이래서 cross entropy를 log loss라고 부르기도 함
- cross entropy가 커지면 log likelihood가 작아지기 때문에, cross entropy의 값을 작게 해야하고, negative log likelihood로 불리기도 함

## 1. 정보량 (Quantity of information)

- 질량이나 높이, 속도와 같은 것들을 단위량을 가지고 측정하고 비교하듯이 정보 또한 그럴 수 있는데, 그 때 등장하는 개념 `정보량`

## 2. Entropy

- 문자열을 출력하는 2개의 기계 'X', 'Y'가 있음
- 기계 X는 A,B,C,D를 각각 0.25의 확률로 출력
- 기계 Y는 A,B,C,D를 각각 0.5,0.125,0.125,0.25 확률로 출력





- 1개의 글자를 기계 Y로 부터 추려내기 위해서는 평균 1.75개 질문, X에서는 2개의 질문을 함
- 만약, 기계가 출력한 100개의 글자를 맞히기 위해서는 기계 Y는 (1.75X100)= 175번, 기계 X는 (2X100) = 200번 질문을 해야함
- 기계 Y가 기계 X보다 더 적은 정보량을 생산함
- `왜냐하면 불확실성(랜덤성)이 더 적기 때문임` : 모든 사건이 같은 확률로 일어나는 것이 가장 불확실함 
- 이러한 불확실성의 측정을 Shannon은 `Entropy` 라 부르고, 이를 `H`라 표시. 단위를 `bit`


- 질문의 횟수는 A는 1번, B와 C는 3번, D는 2번 이는 A,B,C,D의 발생확률과 연관이 있음


- 위의 `가능한 결과의 수`는 `그 사건 발생확률의 역수`임 
- Shannon의 식은 `이산확률분포` 일 때 의 이야기




- Entropy는 `가능한 모든 사건이 같은 확률로 일어날 때 그 최대값을 가짐`
- 각각의 확률이 모두 동등한 상황에서 조금만 벗어나도 entropy는 감소함



- X의 범위에 걸쳐 특정 값들만 높은 왼쪽 분포가 값들끼리 서로 비슷비슷한 오른쪽 분포보다 entropy가 작음
- 즉, `entropy`는 최적의 전략 하에서 그 사건을 예측하는 데 필요한 질문 개수를 의미 
- `최적의 전략 하에서 필요한 질문개수에 대한 기대값`
- entropy가 감소한 다는 것은, 우리가 그 사건을 맞히기 위해서 필요한 질문의 개수가 줄어드는 것을 의미 = 정보량이 줄어든다는 의미
